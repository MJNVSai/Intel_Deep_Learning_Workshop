```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from sklearn.preprocessing import LabelEncoder

from keras.models import Sequential
from keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
train_path = "/content/drive/MyDrive/BTC Dataset-20221029T063940Z-001/BTC Dataset/Training/"

classes = os.listdir(train_path)
```

```python
classes
```

```python
img_data = []
img_labels = []

count = 1

for cls in classes:
  imgs = os.listdir(train_path + cls)

  for i in imgs:
    img_path = train_path + cls + "/" + i
    print(img_path)

    img = image.load_img(img_path,target_size=(224,224,3))
    img = image.img_to_array(img)
    
    img_data.append(img)
    img_labels.append(cls)
    print("\n\n", count, "\n\n")
    count = count + 1
```

```python
X = np.asarray(img_data)
y = np.asarray(img_labels)
```

```python
X.shape, y.shape
```

```python
# Converting Strings to Integers

le = LabelEncoder()
Y = le.fit_transform(y)

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(layers.Conv2D(16, 3, activation = 'relu', input_shape = (224, 224, 3)))
model.add(layers.MaxPooling2D())

model.add(layers.Conv2D(32, 3, activation = 'relu'))
model.add(layers.MaxPooling2D())

model.add(layers.Conv2D(32, 3, activation = 'relu'))
model.add(layers.MaxPooling2D())

model.add(layers.Conv2D(64, 3, activation = 'relu'))
model.add(layers.MaxPooling2D()) 

model.add(layers.Flatten())

model.add(layers.Dense(128,activation = 'relu'))
model.add(layers.Dense(4, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = "adam", metrics = ["accuracy"])
```

```python
model.fit(X_train,y_train,epochs=10, batch_size=32,validation_data=(X_test,y_test))
```

```python
y_pred = np.argmax(model.predict(X_test),axis=-1)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
